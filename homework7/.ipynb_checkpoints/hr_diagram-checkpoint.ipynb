```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
```

```python
print(data.shape)   
# There are 2642 rows (stars) and 2 columns (magnitude an dtemeprature)
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print("Min magnitude:", magnitude.min())
print("Max magnitude:", magnitude.max())

print("Min temperature:", temperature.min())
print("Max temperature:", temperature.max())

#The minimum and maximum values correpsond to stars
#min: index 1988 (lowest magnitude)
#maximum: index 264 (highest magnitude)

#minimimum temperature: star w/ index 1923 (coldest)
#maximum temperature: star w/ index 2613 (hottest)
```

```python
brightest_idx = np.argmin(magnitude)
dimmest_idx = np.argmax(magnitude)

hottest_idx = np.argmax(temperature)
coldest_idx = np.argmin(temperature)

print("Brightest star index:", brightest_idx)
print("Dimmest star index:", dimmest_idx)
print("Hottest star index:", hottest_idx)
print("Coldest star index:", coldest_idx)

#Brightest star is index 1988
#dimmest star is index 264
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, s=20, color="blue", label="Stars")


plt.scatter(log_temperature[brightest_idx], log_magnitude[brightest_idx],
            color="black", marker="*", s=300, label="Brightest Star")

plt.scatter(log_temperature[dimmest_idx], log_magnitude[dimmest_idx],
            color="purple", marker="*", s=300, label="Dimmest Star")

plt.scatter(log_temperature[hottest_idx], log_magnitude[hottest_idx],
            color="green", marker="*", s=300, label="Hottest Star")

plt.scatter(log_temperature[coldest_idx], log_magnitude[coldest_idx],
            color="orange", marker="*", s=300, label="Coldest Star")


plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("HertzsprungRussell Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude (log scale)")

plt.legend()
plt.show()
```
