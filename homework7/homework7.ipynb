```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)
```

```python
x = np.linspace(0, 2 * np.pi, 1000)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
plt.figure(figsize=(25, 5))


linestyles = ['-', '--', '-.', ':', (0, (3, 1, 1, 1))]


plt.subplot(1, 3, 1)
for A, ls in zip(amplitude, linestyles):
    y = make_sine_wave(x, A, 1)  
    plt.plot(x, y, linestyle=ls, label=f"A={A}, w=1")

plt.title("Varying Amplitude", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()


plt.subplot(1, 3, 2)
for w, ls in zip(frequency, linestyles):
    y = make_sine_wave(x, 1, w)  
    plt.plot(x, y, linestyle=ls, label=f"A=1, w={w}")

plt.title("Varying Frequency", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()


plt.subplot(1, 3, 3)
for A, w, ls in zip(amplitude, frequency, linestyles):
    y = make_sine_wave(x, A, w)
    plt.plot(x, y, linestyle=ls, label=f"A={A}, w={w}")

plt.title("Varying Amplitude & Frequency", fontsize=14)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.legend()

plt.show()
```
