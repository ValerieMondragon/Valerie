```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("exoplanet_data.csv.csv", comment = "#")
```

```python
print("Shape of the dataset:", df.shape)
```

```python
rows, cols = df.shape
print("Number of rows:", rows)
print("Number of columns:", cols)
```

```python
print("\nColumn titles:")
for col in df.columns:
    print(col)
```

```python
print("\nColumns with index numbers:")
for i, col in enumerate(df.columns):
    print(i, "-", col)
```

```python
# Table Label: Planet Name
# Database Column Name: pl_name
# NASA Description: The official name of the exoplanet.
# My explanation: literally just the planet’s name (Kepler-22 b type stuff).

# Table Label: Host Name
# Database Column Name: hostname
# NASA Description: The name of the star the planet orbits.
# My explanation: the parent star of the planet.

# Table Label: Discovery Method
# Database Column Name: discoverymethod
# NASA Description: The technique used to detect the planet 
# (transit, radial velocity, direct imaging, etc).
# My explanation: how we found the planet in the first place.

# Table Label: Spectral Type
# Database Column Name: st_spectype
# NASA Description: The spectral class of the host star 
# (like G2V for a Sun-like star, M-type for cool red star, etc).
# My explanation: tells you what kind of star it is, and kinda the vibe (hot, cool, sun-like, etc).

# Table Label: Equilibrium Temperature [K]
# Database Column Name: pl_eqt
# NASA Description: Estimated planetary temperature in Kelvin assuming no atmosphere.
# My explanation: sorta the “expected” temperature the planet should have.

# Table Label: Planet Mass or Mass*sin(i) [Jupiter Mass]
# Database Column Name: pl_bmassj
# NASA Description: Planet mass expressed in Jupiter masses (or minimum mass).
# My explanation: how big the planet is compared to Jupiter. Values >1 = huge planet.
```

```python
reduced_cols = [
    "pl_name",
    "hostname",
    "discoverymethod",
    "st_spectype",
    "pl_eqt",
    "pl_bmassj"
]

df_small = df[reduced_cols]
df_small.head()
```

```python
df_small_clean = df_small.dropna()
df_small_clean.shape
```

```python
rows, cols = df_small_clean.shape

print("Shape:", df_small_clean.shape)
print("Rows:", rows)
print("Columns:", cols)

print("\nFirst five rows:")
print(df_small_clean.head())

print("\nLast five rows:")
print(df_small_clean.tail())

print("\nColumn titles:")
print(list(df_small_clean.columns))
```

```python
df_small_clean
```

```python
method_counts = df_small_clean["discoverymethod"].value_counts()
method_counts
```

```python
#These are the discovery methods in my filtered dataset. Transit totally dominates (no surprise
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df_small_clean,
    x="pl_bmassj",
    y="pl_eqt",
    hue="discoverymethod",
    alpha=0.8
)

plt.title("Planet Mass vs Equilibrium Temperature")
plt.xlabel("Planet Mass (Jupiter Masses)")
plt.ylabel("Equilibrium Temperature (K)")
plt.legend(title="Discovery Method")
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(
    data=df_small_clean,
    x=np.log(df_small_clean["pl_bmassj"]),
    y=np.log(df_small_clean["pl_eqt"]),
    hue="discoverymethod",
    alpha=0.7
)

plt.title("Log-Log: Mass vs Equilibrium Temp")
plt.xlabel("log(Mass)")
plt.ylabel("log(Equilibrium Temperature)")

# required note:
# Using log scales here because the original values are super skewed.

plt.legend(title="Discovery Method")
plt.show()

# Imaging on the plot is showing up on the right of the mass axis. Direct imaging only works for huge planets that are really bright and really far from their stars. Smaller planets are basically impossible to photograph directly, so imaging is basically the “giant, cold planets” category.
```

```python
big_planets = df_small_clean[df_small_clean["pl_bmassj"] > 5]

coldest_idx = big_planets["pl_eqt"].idxmin()
coldest_planet = big_planets.loc[coldest_idx]

coldest_planet
```
