```python
arr = np.array([
    [2,3,5],
    [14,6,8],
    [11,13,17],
    [7,10,13]
])

def is_prime(n):
    if n < 2:
        return False 
    for i in range (2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def rows_with_prime(a):
    rows_list = []
    for row in a:
        has_prime = False
        for value in row:
            if is_prime(value):
                has_prime = True
                break
        if has_prime: 
            rows_list.append(row)
    return np.array(rows_list)

result = rows_with_prime(arr)
print(result)
```
